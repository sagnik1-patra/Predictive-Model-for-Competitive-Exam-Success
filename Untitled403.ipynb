```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import save_model

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Predictive Model for Competitive Exam Success"
DATA_PATH = os.path.join(BASE_DIR, "RS_Session_256_AU_2981_A_to_F.csv")

MODEL_DIR = os.path.join(BASE_DIR, "models")
ARTIFACT_DIR = os.path.join(BASE_DIR, "artifacts")
OUTPUT_DIR = os.path.join(BASE_DIR, "outputs")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")
CONFIG_DIR = os.path.join(BASE_DIR, "configs")

for d in [MODEL_DIR, ARTIFACT_DIR, OUTPUT_DIR, GRAPH_DIR, CONFIG_DIR]:
    os.makedirs(d, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df = shuffle(df, random_state=42)

df_numeric = df.select_dtypes(include=[np.number])
X = df_numeric.iloc[:, :-1]
y = df_numeric.iloc[:, -1]

print("Target distribution:")
print(y.value_counts())

# ============================================================
# HEATMAP
# ============================================================

corr = df_numeric.corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.title("Feature Correlation Heatmap")
plt.show()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.title("Feature Correlation Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "bis_heatmap.png"))
plt.close()

# ============================================================
# SCALING
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, os.path.join(ARTIFACT_DIR, "bis_scaler.pkl"))

# ============================================================
# HYBRID AIS + BAT ALGORITHM FEATURE OPTIMIZATION
# ============================================================

def hybrid_ais_ba(X, y, bats=20, iterations=25):
    n_features = X.shape[1]

    # AIS population (antibodies)
    positions = np.random.rand(bats, n_features)
    velocities = np.zeros((bats, n_features))
    frequencies = np.random.rand(bats)
    loudness = np.ones(bats)
    pulse_rate = np.random.rand(bats)

    fitness = np.zeros(bats)
    best_solution = None
    best_score = -1

    def evaluate(weights):
        Xw = X * weights
        X_tr, X_te, y_tr, y_te = train_test_split(
            Xw, y, test_size=0.2, random_state=42
        )
        model = GradientBoostingClassifier()
        model.fit(X_tr, y_tr)
        preds = model.predict(X_te)
        return accuracy_score(y_te, preds)

    # Initial AIS evaluation
    for i in range(bats):
        fitness[i] = evaluate(positions[i])
        if fitness[i] > best_score:
            best_score = fitness[i]
            best_solution = positions[i].copy()

    # BAT refinement
    f_min, f_max = 0, 2
    alpha, gamma = 0.9, 0.9

    for _ in range(iterations):
        for i in range(bats):
            frequencies[i] = f_min + (f_max - f_min) * np.random.rand()
            velocities[i] += (positions[i] - best_solution) * frequencies[i]
            new_pos = positions[i] + velocities[i]

            # Local search (BA pulse rate)
            if np.random.rand() > pulse_rate[i]:
                new_pos = best_solution + 0.001 * np.random.randn(n_features)

            new_pos = np.clip(new_pos, 0, 1)
            new_score = evaluate(new_pos)

            if (new_score > fitness[i]) and (np.random.rand() < loudness[i]):
                positions[i] = new_pos
                fitness[i] = new_score
                loudness[i] *= alpha
                pulse_rate[i] = pulse_rate[i] * (1 - np.exp(-gamma))

                if new_score > best_score:
                    best_solution = new_pos.copy()
                    best_score = new_score

    return best_solution

bis_weights = hybrid_ais_ba(X_scaled, y)
X_bis = X_scaled * bis_weights

# ============================================================
# TRAIN-TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_bis, y, test_size=0.2, random_state=42
)

# ============================================================
# MODEL COMPARISON
# ============================================================

models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(),
    "Hybrid AIS + BA (GB)": GradientBoostingClassifier(
        n_estimators=150, learning_rate=0.05, max_depth=3
    )
}

accuracies = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    accuracies[name] = accuracy_score(y_test, preds)

# ============================================================
# COMPARISON GRAPH
# ============================================================

plt.figure()
plt.bar(accuracies.keys(), accuracies.values())
plt.title("BIS Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.show()

plt.figure()
plt.bar(accuracies.keys(), accuracies.values())
plt.title("BIS Model Accuracy Comparison")
plt.savefig(os.path.join(GRAPH_DIR, "bis_comparison_graph.png"))
plt.close()

# ============================================================
# FINAL MODEL
# ============================================================

final_model = models["Hybrid AIS + BA (GB)"]
joblib.dump(final_model, os.path.join(MODEL_DIR, "bis_gb_ais_ba_model.pkl"))

# ============================================================
# ACCURACY GRAPH
# ============================================================

train_acc = accuracy_score(y_train, final_model.predict(X_train))
test_acc = accuracy_score(y_test, final_model.predict(X_test))

plt.figure()
plt.plot(["Train", "Test"], [train_acc, test_acc], marker="o")
plt.title("BIS Train vs Test Accuracy")
plt.show()

plt.figure()
plt.plot(["Train", "Test"], [train_acc, test_acc], marker="o")
plt.title("BIS Train vs Test Accuracy")
plt.savefig(os.path.join(GRAPH_DIR, "bis_accuracy_graph.png"))
plt.close()

# ============================================================
# PREDICTION + GRAPH
# ============================================================

y_prob = final_model.predict_proba(X_test)[:, 1]

plt.figure()
plt.plot(y_prob)
plt.title("BIS Selection Probability")
plt.xlabel("Students")
plt.ylabel("Probability")
plt.show()

plt.figure()
plt.plot(y_prob)
plt.title("BIS Selection Probability")
plt.savefig(os.path.join(GRAPH_DIR, "bis_prediction_graph.png"))
plt.close()

# ============================================================
# SAVE RESULTS
# ============================================================

results = pd.DataFrame(X_test, columns=X.columns)
results["Actual"] = y_test.values
results["Predicted_Probability"] = y_prob

results.to_csv(
    os.path.join(OUTPUT_DIR, "bis_final_results.csv"),
    index=False
)

with open(os.path.join(OUTPUT_DIR, "bis_predictions.json"), "w") as f:
    json.dump({"selection_probability": y_prob.tolist()}, f, indent=4)

# ============================================================
# SAVE NEURAL MODEL (.h5)
# ============================================================

nn = Sequential([
    Dense(32, activation="relu", input_shape=(X_train.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")
])

nn.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
nn.fit(X_train, y_train, epochs=10, batch_size=16, verbose=0)

save_model(nn, os.path.join(MODEL_DIR, "bis_selection_model.h5"))

config = {
    "project": "Predictive Model for Competitive Exam Success",
    "model": "Hybrid AIS + Bat Algorithm + Gradient Boosting",
    "accuracy": float(test_acc),
    "features": list(X.columns)
}

with open(os.path.join(CONFIG_DIR, "bis_model_config.yaml"), "w") as f:
    yaml.dump(config, f)

print("\n BIS (HYBRID AIS + BA) PIPELINE COMPLETED SUCCESSFULLY")
```
