```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import save_model

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Predictive Model for Competitive Exam Success"
DATA_PATH = os.path.join(BASE_DIR, "RS_Session_256_AU_2981_A_to_F.csv")

MODEL_DIR = os.path.join(BASE_DIR, "models")
ARTIFACT_DIR = os.path.join(BASE_DIR, "artifacts")
OUTPUT_DIR = os.path.join(BASE_DIR, "outputs")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")
CONFIG_DIR = os.path.join(BASE_DIR, "configs")

for d in [MODEL_DIR, ARTIFACT_DIR, OUTPUT_DIR, GRAPH_DIR, CONFIG_DIR]:
    os.makedirs(d, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df = shuffle(df, random_state=42)

df_numeric = df.select_dtypes(include=[np.number])
X = df_numeric.iloc[:, :-1]
y = df_numeric.iloc[:, -1]

print("Target Distribution:")
print(y.value_counts())

# ============================================================
# HEATMAP
# ============================================================

corr = df_numeric.corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.title("Feature Correlation Heatmap")
plt.show()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.title("Feature Correlation Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "hybrid_heatmap.png"))
plt.close()

# ============================================================
# SCALING
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, os.path.join(ARTIFACT_DIR, "hybrid_scaler.pkl"))

# ============================================================
# HYBRID AIS + CSA FEATURE OPTIMIZATION
# ============================================================

def hybrid_ais_csa(X, y, population=20, iterations=20, awareness=0.1):
    n_features = X.shape[1]
    memory = np.random.rand(population, n_features)
    fitness = np.zeros(population)

    def evaluate(weights):
        Xw = X * weights
        X_tr, X_te, y_tr, y_te = train_test_split(
            Xw, y, test_size=0.2, random_state=42
        )
        model = GradientBoostingClassifier()
        model.fit(X_tr, y_tr)
        preds = model.predict(X_te)
        return accuracy_score(y_te, preds)

    # Initial AIS evaluation
    for i in range(population):
        fitness[i] = evaluate(memory[i])

    # CSA refinement
    for _ in range(iterations):
        for i in range(population):
            if np.random.rand() > awareness:
                j = np.random.randint(population)
                step = np.random.rand()
                new_pos = memory[i] + step * (memory[j] - memory[i])
            else:
                new_pos = np.random.rand(n_features)

            new_pos = np.clip(new_pos, 0, 1)
            new_score = evaluate(new_pos)

            if new_score > fitness[i]:
                memory[i] = new_pos
                fitness[i] = new_score

    best_idx = np.argmax(fitness)
    return memory[best_idx]

hybrid_weights = hybrid_ais_csa(X_scaled, y)
X_hybrid = X_scaled * hybrid_weights

# ============================================================
# TRAIN-TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_hybrid, y, test_size=0.2, random_state=42
)

# ============================================================
# MODEL COMPARISON
# ============================================================

models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(),
    "Hybrid AIS + CSA (GB)": GradientBoostingClassifier(
        n_estimators=150, learning_rate=0.05, max_depth=3
    )
}

accuracies = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    accuracies[name] = accuracy_score(y_test, preds)

# ============================================================
# COMPARISON GRAPH
# ============================================================

plt.figure()
plt.bar(accuracies.keys(), accuracies.values())
plt.title("Hybrid Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.show()

plt.figure()
plt.bar(accuracies.keys(), accuracies.values())
plt.title("Hybrid Model Accuracy Comparison")
plt.savefig(os.path.join(GRAPH_DIR, "hybrid_comparison_graph.png"))
plt.close()

# ============================================================
# FINAL MODEL
# ============================================================

final_model = models["Hybrid AIS + CSA (GB)"]
joblib.dump(final_model, os.path.join(MODEL_DIR, "hybrid_gb_ais_csa_model.pkl"))

# ============================================================
# ACCURACY GRAPH
# ============================================================

train_acc = accuracy_score(y_train, final_model.predict(X_train))
test_acc = accuracy_score(y_test, final_model.predict(X_test))

plt.figure()
plt.plot(["Train", "Test"], [train_acc, test_acc], marker="o")
plt.title("Hybrid Train vs Test Accuracy")
plt.show()

plt.figure()
plt.plot(["Train", "Test"], [train_acc, test_acc], marker="o")
plt.title("Hybrid Train vs Test Accuracy")
plt.savefig(os.path.join(GRAPH_DIR, "hybrid_accuracy_graph.png"))
plt.close()

# ============================================================
# PREDICTION + GRAPH
# ============================================================

y_prob = final_model.predict_proba(X_test)[:, 1]

plt.figure()
plt.plot(y_prob)
plt.title("Hybrid Selection Probability")
plt.xlabel("Students")
plt.ylabel("Probability")
plt.show()

plt.figure()
plt.plot(y_prob)
plt.title("Hybrid Selection Probability")
plt.savefig(os.path.join(GRAPH_DIR, "hybrid_prediction_graph.png"))
plt.close()

# ============================================================
# SAVE RESULTS
# ============================================================

results = pd.DataFrame(X_test, columns=X.columns)
results["Actual"] = y_test.values
results["Predicted_Probability"] = y_prob

results.to_csv(
    os.path.join(OUTPUT_DIR, "hybrid_final_results.csv"),
    index=False
)

with open(os.path.join(OUTPUT_DIR, "hybrid_predictions.json"), "w") as f:
    json.dump({"selection_probability": y_prob.tolist()}, f, indent=4)

# ============================================================
# SAVE NEURAL MODEL (.h5)
# ============================================================

nn = Sequential([
    Dense(32, activation="relu", input_shape=(X_train.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1, activation="sigmoid")
])

nn.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
nn.fit(X_train, y_train, epochs=10, batch_size=16, verbose=0)

save_model(nn, os.path.join(MODEL_DIR, "hybrid_selection_model.h5"))

# ============================================================
# SAVE CONFIG YAML
# ============================================================

config = {
    "project": "Predictive Model for Competitive Exam Success",
    "model": "Hybrid AIS + CSA + Gradient Boosting",
    "accuracy": float(test_acc),
    "features": list(X.columns)
}

with open(os.path.join(CONFIG_DIR, "hybrid_model_config.yaml"), "w") as f:
    yaml.dump(config, f)

print("\n HYBRID AIS + CSA PIPELINE COMPLETED SUCCESSFULLY")
```
